# Retention analysis
Compute weekly retention from Athena.

```python
import pandas as pd
import awswrangler as wr
from datetime import datetime
```

```python
BUCKET = 's3://analytics-bucket/retention/weekly.parquet'
df = pd.read_parquet(BUCKET)
```

```python
query = """
SELECT user_id, week, retained
FROM analytics.retention
WHERE week >= '2026-01-01'
"""
df2 = wr.athena.read_sql_query(query, database='analytics')
```

```python
def weekly_mean(frame):
    return frame.groupby('week')['retained'].mean()

weekly_mean(df2).plot()
```
